```python
import numpy as np
import matplotlib.pyplot as plt
n=200

def visualizeScatter(X,Y):
  plt.scatter(X,Y)
  plt.title("Data")
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.show()

def visualizeRe(X,Y,yhat):
  fig = plt.figure(figsize=(8,6))
  plt.plot(X,Y,'b.')
  plt.plot(X,yhat,'g*')
  plt.show()


def getData():
    X = np.random.randn(n,1)
    Y = 2*X + 1 + np.random.randn(n,1)
    return X, Y

def calculateW(X,Y):
  temp1 = np.linalg.inv(np.dot(X.T, X))
  temp2 = np.dot(X.T,Y)
  W = np.dot(temp1,temp2)
  return W

X,Y = getData()
O = np.ones((n,1))
X1 = np.append(X,O,axis=1)
print(X.shape, O.shape, X1.shape,Y.shape)
visualizeScatter(X,Y)

w = calculateW(X1,Y)
print(w) 
yhat = np.dot(X1,w)
visualizeRe(X,Y,yhat)

```
